```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([1, 2, 3, 4, 5])
y_train = np.array([2, 3, 5, 7, 11])
```

```python
plt.scatter(x_train, y_train, color='blue', label='Training Data')
plt.title('Training Data Scatter Plot')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#implement linear regression

def comput_LR(x_train, y_train, w , b):
    
    m = len(x_train)
    f = np.zeros(m)

    for i in range(m):
        f[i] = w * x_train[i] + b
        
    return f
```

```python
#test comput_LR
w = 2
b = -3
f = comput_LR(x_train, y_train, w, b)
plt.scatter(x_train, y_train, color='blue', label='Training Data')
plt.plot(x_train, f, color='red', label='Linear Regression Line')
plt.title('Linear Regression Fit')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def cost(x_train, y_train, w, b):
    m = len(x_train)
    cost = 0.0

    for i in range(m):
        # calculate predicted value
        y_hat = w * x_train[i] + b

        # accumulate squared error
        cost += (y_hat - y_train[i]) ** 2

    # calculate average cost
    total_cost = (1 / (2 * m)) * cost

    return total_cost
```

```python
#test cost function
w = 2
b = -3  
x = cost(x_train, y_train, w, b)
print("Cost:", str(x))
```
